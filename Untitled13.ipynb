Notebook ini mendokumentasikan seluruh tahapan permodelan machine learning mulai dari pembersihan data, feature enginering, penyeimbangan data, tuning hyperparameter, ensemble model, hingga evaluasi model dengan cross validation.

```python
# =====================================================
# ðŸ“˜ IMPORT LIBRARY
# =====================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    confusion_matrix, roc_auc_score, roc_curve,
    classification_report, accuracy_score,
    precision_score, recall_score, f1_score
)
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, VotingClassifier
from imblearn.over_sampling import SMOTE
```

library di atas digunakan untuk melakukan analisis data, visualisasi, preprocessing, training model, dan evaluasi performa, SMOTE digunakan untuk menangani ketidak seimbangan kelas pada data target

```python
# =====================================================
#  1. BACA DATA
# =====================================================
train = pd.read_csv('/content/data_training.csv')
test = pd.read_csv('/content/data_testing.csv')

print("Data training:", train.shape)
print("Data testing:", test.shape)
print(train.head())
```

Dataset dibagii menjadi data training (untuk melatih model) dan data testing (untuk prediksi akhir). lanhkag ini memastikan pemisahan data sesuai ketentuan.

```python
# =====================================================
#  2. PEMBERSIHAN DATA
# =====================================================
print("\nCek Missing Values:")
print(train.isnull().sum())

X = train.drop(['Id', 'quality'], axis=1)
y = train['quality']
```

Tahap ini dilakukan untuk memasstikan tidak ada missing value, Kolom id dihapus karena hanya bersifat identitas, dan quality digunakan sebagai label target klasifikasi.


```python
# =====================================================
#  HAPUS FITUR TIDAK INFORMATIF
# =====================================================
# Hapus kolom dengan hanya 1 nilai unik
constant_features = [col for col in X.columns if X[col].nunique() == 1]
X.drop(columns=constant_features, inplace=True)

# Hapus fitur dengan korelasi tinggi (>0.9)
corr_matrix = X.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
high_corr_features = [column for column in upper.columns if any(upper[column] > 0.9)]
X.drop(columns=high_corr_features, inplace=True)

print(" Fitur dihapus:", constant_features + high_corr_features)

# Terapkan penghapusan kolom yang sama pada data test
X_test = test.drop(['Id'], axis=1)
X_test.drop(columns=[col for col in constant_features + high_corr_features if col in X_test.columns],
            inplace=True, errors='ignore')
```

Langkah ini bertujuan untuk menghapus fitur yang tidak memberikan informasi baru, seperti:


1.   Fitur dengan hanya 1 nilai unik (tidak ada variasi)
2.   Fitur dengan korelasi >0.9
hal ini membantu mengurangi multikolinearitas dan membuat model lebih efisien.



```python
# =====================================================
#  FEATURE ENGINEERING TAMBAHAN (TRAIN & TEST)
# =====================================================
def add_features(df):
    if all(f in df.columns for f in ['fixed acidity', 'volatile acidity']):
        df['acidity_ratio'] = df['fixed acidity'] / (df['volatile acidity'] + 1e-6)
    if all(f in df.columns for f in ['density', 'alcohol']):
        df['density_alcohol_ratio'] = df['density'] / (df['alcohol'] + 1e-6)
    if all(f in df.columns for f in ['fixed acidity', 'citric acid', 'volatile acidity']):
        df['total_acidity'] = df['fixed acidity'] + df['citric acid'] + df['volatile acidity']
    return df

X = add_features(X)
X_test = add_features(X_test)
```

Menambahkan fitur turunan yang dapat memberikan informasi kimia tambahan, misalnya rasio keasaman dan rasio densitas terhadap alkohol.
Langkah ini membantu model memahami hubungan non-linear antar variabel.

```python
# =====================================================
#  3. STANDARISASI DATA
# =====================================================
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_test_scaled = scaler.transform(X_test)
```

Proses standardisasi mengubah data menjadi skala yang sama (rata-rata 0 dan deviasi standar 1) agar model seperti Random Forest dan Gradient Boosting bekerja lebih optimal.

```python
# =====================================================
#  4. PEMBAGIAN DATA TRAIN & VALIDASI
# =====================================================
X_train, X_val, y_train, y_val = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
```

Data dibagi menjadi 80% train dan 20% validasi.
Kemudian, dilakukan SMOTE (Synthetic Minority Oversampling Technique) untuk menyeimbangkan distribusi kelas agar model tidak bias terhadap kelas mayoritas.

```python
# =====================================================
# 5. PENYEIMBANGAN DATA (SMOTE HANYA PADA TRAINING)
# =====================================================
print("\nDistribusi kelas sebelum SMOTE:")
print(y_train.value_counts())

sm = SMOTE(random_state=42, k_neighbors=3)
X_train_res, y_train_res = sm.fit_resample(X_train, y_train)

print("\nDistribusi kelas setelah SMOTE:")
print(y_train_res.value_counts())
```

```python
# =====================================================
# 6. GRID SEARCH UNTUK RANDOM FOREST (DENGAN CLASS_WEIGHT)
# =====================================================
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'class_weight': ['balanced']
}

rf = RandomForestClassifier(random_state=42)

grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=5,
    n_jobs=-1,
    scoring='accuracy',
    verbose=1
)

grid_search.fit(X_train_res, y_train_res)

print("\n Parameter terbaik hasil GridSearch:")
print(grid_search.best_params_)

best_rf = grid_search.best_estimator_
```

Menggunakan GridSearchCV untuk mencari kombinasi parameter terbaik bagi model Random Forest.
Hal ini meningkatkan akurasi model dengan menemukan parameter optimal secara sistematis.

```python
# =====================================================
#  6. ENSEMBLE MODEL (VOTING)
# =====================================================
gb = GradientBoostingClassifier(random_state=42)
ensemble = VotingClassifier(
    estimators=[('rf', best_rf), ('gb', gb)],
    voting='soft'
)

ensemble.fit(X_train_res, y_train_res)
```

Model Voting Classifier menggabungkan dua algoritma kuat (Random Forest dan Gradient Boosting) dengan metode soft voting, sehingga meningkatkan stabilitas dan akurasi prediksi.

```python
# =====================================================
#  7. EVALUASI MODEL
# =====================================================
y_pred = ensemble.predict(X_val)
y_proba = ensemble.predict_proba(X_val)

cm = confusion_matrix(y_val, y_pred)
auc_score = roc_auc_score(y_val, y_proba, multi_class='ovr')

print("\n================ ENSEMBLE MODEL (RF + GB) ================")
print("Confusion Matrix:\n", cm)
print("\nClassification Report:\n", classification_report(y_val, y_pred))
print("AUC Score:", round(auc_score, 4))
```

1. Confusion Matrix: menunjukkan distribusi prediksi benar dan salah.

2. Classification Report: berisi precision, recall, f1-score.

3. AUC Score: mengukur kemampuan model membedakan antar kelas (semakin tinggi semakin baik).

```python
# =====================================================
#  8. VISUALISASI CONFUSION MATRIX
# =====================================================
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix - Ensemble (RF + GB)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.show()
```

Visualisasi membantu mengevaluasi seberapa baik model mengenali setiap kelas dan menampilkan area ROC untuk setiap kategori target.

```python
# =====================================================
#  9. ROC CURVE
# =====================================================
fpr, tpr = {}, {}
for i, cls in enumerate(np.unique(y_val)):
    fpr[cls], tpr[cls], _ = roc_curve((y_val == cls).astype(int), y_proba[:, i])
    plt.plot(fpr[cls], tpr[cls], label=f"Class {cls}")

plt.plot([0,1],[0,1],'k--')
plt.title("ROC Curve - Ensemble (RF + GB)")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# =====================================================
# 10. NILAI EVALUASI
# =====================================================
acc = accuracy_score(y_val, y_pred)
prec = precision_score(y_val, y_pred, average='macro')
rec = recall_score(y_val, y_pred, average='macro')
f1 = f1_score(y_val, y_pred, average='macro')

metrics = {
    'Accuracy': acc,
    'Precision': prec,
    'Recall': rec,
    'F1-Score': f1,
    'AUC': auc_score
}

plt.figure(figsize=(7,5))
sns.barplot(x=list(metrics.keys()), y=list(metrics.values()), palette='viridis')
plt.title("Evaluasi Model Ensemble (RF + GB)")
plt.ylabel("Score")
plt.ylim(0, 1)

for i, v in enumerate(metrics.values()):
    plt.text(i, v + 0.01, f"{v:.3f}\n({v*100:.1f}%)",
             ha='center', va='bottom', fontweight='bold', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# =====================================================
#  11. FEATURE IMPORTANCE (DARI RF)
# =====================================================
feature_importances = pd.Series(best_rf.feature_importances_, index=X.columns)
feature_importances = feature_importances.sort_values(ascending=False)

plt.figure(figsize=(8,6))
sns.barplot(x=feature_importances, y=feature_importances.index, palette='crest')
plt.title("Feature Importance - Random Forest (Best Estimator)")
plt.xlabel("Importance Score")
plt.ylabel("Features")
plt.tight_layout()
plt.show()

print("\nFitur dengan pengaruh terbesar pada prediksi kualitas:")
print(feature_importances.head(10))
```

```python
# =====================================================
#  12. VALIDASI DENGAN STRATIFIED K-FOLD
# =====================================================
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(ensemble, X_scaled, y, cv=cv, scoring='accuracy')
print(f"\nRata-rata akurasi dengan 5-Fold CV: {cv_scores.mean():.4f}")
```

```python
# =====================================================
#  13. PREDIKSI DATA TESTING
# =====================================================
prediksi_quality = ensemble.predict(X_test_scaled)

hasil_prediksi = pd.DataFrame({
    'id': test['Id'],
    'quality': prediksi_quality
})

hasil_prediksi.to_csv('hasilprediksi_123.csv', index=False)
print("\nâœ… File hasilprediksi_123.csv berhasil dibuat!")
hasil_prediksi.head()
```
